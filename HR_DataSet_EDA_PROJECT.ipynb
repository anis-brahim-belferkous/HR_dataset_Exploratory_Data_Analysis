# Exploratory Data analysis on a HRdataset



**HR data**  from a reputed organization provides valuable insights into the workforce and human resources management practices. 

Through the analysis of this data, organization can gain a deeper understanding of their employees, their performance, and the factors influencing their engagement and satisfaction.

By delving into the HR data through EDA, we aim to gain valuable insights that can inform decision-making processes related to talent acquisition, performance evaluation, employee retention, and overall HR strategy. These insights can help the organization optimize its human resources practices, foster a positive work environment, and enhance overall employee well-being and productivity. 






---



##What is Exploratory data analysis(EDA) ?
 In statistics, exploratory data analysis is an approach to analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods. A statistical model can be used or not, but primarily EDA is for seeing what the data can tell us beyond the formal modeling or hypothesis testing task.

It is understanding the data sets by summarizing their main characteristics often plotting them visually. This step is very important especially when we arrive at modeling the data in order to apply Machine learning. Plotting in EDA consists of Histograms, Box plot, Scatter plot and many more. It often takes much time to explore the data. Through the process of EDA, we can ask to define the problem statement or definition on our data set which is very important

## Number of Employees by Position and Department

````python
position_counts = df['Position'].value_counts()
print(position_counts)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Group the data by 'Department', 'Position', and 'Sex' and get the count
position_gender_counts = S_df.groupby(['Department', 'Position', 'Sex']).size().unstack().reset_index()

# Plot the data
plt.figure(figsize=(20, 12))

# Loop through the data and plot the bars with count labels
for i, (position, male_count, female_count) in enumerate(zip(position_gender_counts['Position'], position_gender_counts['Male'], position_gender_counts['Female'])):
    plt.bar(position, male_count, label='Male', color='blue')
    plt.bar(position, female_count, label='Female', bottom=male_count, color='pink')
    
    # Add count labels for both male and female counts
    if np.isfinite(male_count):
        plt.text(i, male_count/2, f"{male_count}", ha='center', va='center')
    if np.isfinite(male_count + female_count):
        plt.text(i, male_count + female_count/2, f"{female_count}", ha='center', va='center')

# Customize the plot
plt.xlabel('Position')
plt.ylabel('Count')
plt.title('Number of Employees by Position and Department (Separated by Gender)')
plt.xticks(rotation=90)

# Adjust the legend
plt.legend(['Male', 'Female'], title='Gender', loc='upper right', bbox_to_anchor=(1.0, 1))

plt.tight_layout()

# Show the plot
plt.show()
````

# Termination Reason Distribution

````python


# Filter the dataframe to exclude "N/A-StillEmployed" term reason
filtered_df = S_df[~(S_df['TermReason'] == 'N/A-StillEmployed')]

# Count the frequency of each TermReason
term_reason_counts = filtered_df['TermReason'].value_counts()
mylabels = term_reason_counts.index




fig1, ax1 = plt.subplots(figsize=(20, 8))
ax1.pie(term_reason_counts,labels = mylabels ,autopct='%1.1f%%',startangle=180)

#draw circle
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.title("Termination Reason Distribution", fontsize=16, fontweight='bold', pad=20)

plt.legend(title="Status:", loc='upper right')
plt.tight_layout();
````

# Correlation between continuous variables in Dataset

````python

fig = plt.figure(figsize=(20, 8))
g = sns.heatmap(S_df.corr(), cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

# Add positive and negative correlation axes
g.spines['top'].set_visible(True)
g.spines['bottom'].set_visible(True)
g.spines['left'].set_visible(True)
g.spines['right'].set_visible(True)

# Customize the axis labels
g.set_xticklabels(g.get_xticklabels(), rotation=45)
g.set_yticklabels(g.get_yticklabels(), rotation=0)

# Set the title
g.set_title('Correlation between continuous variables')

plt.show()
````
